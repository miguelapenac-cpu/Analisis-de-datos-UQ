#9. ¿Cuántas personas hay ocupadas en el departamento del Tolima , cuyas edades están entre los 20 y los 25 años , qué porcentaje de la población total del Tolima representan ? 

```python
import pandas as pd

# Define the URL for the 'Ocupados' dataset
url_ocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV'
# Define the URL for the 'CGSSE' dataset (needed for P6040 and AREA)
url_cgsse ='https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/cgsse.CSV'

# Load DataFrames
df_cgsse = pd.read_csv(url_cgsse, encoding='latin-1', sep=';', low_memory=False)
df_ocupados = pd.read_csv(url_ocupados, encoding='latin-1', sep=';', low_memory=False)

# Prepare df_cgsse subset for merging, renaming 'AREA' to 'DPTO_TEMP' to avoid column collision
df_cgsse_for_merge = df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P6040', 'AREA']].copy()
df_cgsse_for_merge.rename(columns={'AREA': 'DPTO_TEMP'}, inplace=True)

# Merge df_ocupados with the prepared df_cgsse subset to get P6040 (age) and DPTO_TEMP (department) columns
df_ocupados = pd.merge(df_ocupados, df_cgsse_for_merge,
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')

# 1. Calcular cantidad de personas ocupadas en el Tolima entre 20 y 25 años
ocupados_tolima_20_25 = df_ocupados[(df_ocupados['OCI']==1) & 
                                   (df_ocupados['P6040']>=20) & 
                                   (df_ocupados['P6040']<=25) & 
                                   (df_ocupados['DPTO_TEMP']==73)]['FEX_C18'].sum()

print(f'Personas ocupadas en el Tolima entre los 20 y 25 años: {ocupados_tolima_20_25:,.2f}')

# 2. Calcular la población total del Tolima using df_cgsse's AREA column
poblacion_total_tolima = df_cgsse[df_cgsse['AREA']==73]['FEX_C18'].sum()

# 3. Calcular el porcentaje de personas ocupadas (20-25 años) sobre la población total del Tolima
porcentaje_ocupados_tolima = (ocupados_tolima_20_25 / poblacion_total_tolima) * 100

print(f'Representan el {porcentaje_ocupados_tolima:,.2f}% de la población total del Tolima.')
```

# Respuesta 9

Hay 23575 personas ocupadas en el Tolima y el porcentaje de la poblacion ocupada entre las edades de 20 y 25 es de 4.62%
